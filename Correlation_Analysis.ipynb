```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr

# Load data
df = pd.read_csv("defence_pressure_vs_goals_conceded.csv")

# Preview
print(df.head())

# Drop missing values
df = df.dropna()

# Correlation
corr = df['avg_defence_pressure'].corr(df['avg_goals_conceded'])
print(f"Correlation: {corr:.3f}")

# Pearson test
r, p_value = pearsonr(
    df['avg_defence_pressure'],
    df['avg_goals_conceded']
)

print(f"Pearson r: {r:.3f}")
print(f"P-value: {p_value:.5f}")
```

```python
plt.figure(figsize=(8,6))
sns.regplot(
    x='avg_defence_pressure',
    y='avg_goals_conceded',
    data=df,
    scatter_kws={'alpha':0.6}
)

plt.title("Defence Pressure vs Average Goals Conceded")
plt.xlabel("Average Defence Pressure")
plt.ylabel("Average Goals Conceded")
plt.grid(True)
plt.show()
```
